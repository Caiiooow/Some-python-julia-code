```julia



# Taking for example the approximation of the cosine function, we use the taylor series 
   # Of course, using the small angle approximation 

# Let's write a function that takes a whole vector as input and one by one evaluates the values by the approx
```

```julia
function approxcos(x)
    # Initilising the ouput, note the use of size() to specify the dimensions of the output
    outval = Array{Float64}(undef, size(x))
    
    
    # Now we shall loop over the input vector
    
    
    ii = 0 # Index into the vector since it is 1D
    
    for aa in x
        y = 1 - aa^2/2 + aa^4/24 - aa^6/720 + aa^8/(56*720)  # By Taylor Series
        ii += 1
        
        outval[ii] = y  # This stores the approximation 
    end
    return outval
end



x1 = 4*rand(10) # This will perform  to fill uniformily spaces between [0,1)

# Here we use it for plotting values on the interval [0, 4)

x2 = range(0., stop = 4, step = 0.01)

y1 = approxcos(x1)
y2 = cos.(x2);
```

```julia
using Plots; gr() # Make it more easy to put it all on the same line 

scatter(x1, y1, legend=:false, title = "6th Order Approximation to Cos(x)" )

plot!(x2,y2, color = "black") # Adding the accurate line with plot!()
```
